2. Crear una función que reciba como entrada un número natural y verifique si es primo : "Es primo" , "No es primo"

```python
def prim(x): # función
    i=x # Variable
    c=0
    while(i>=1): 
        a=x%i
        print(a)
        i-=1
        if a == 0 :
            c+=1
    if c == 2:
        print("Primo")
    else:
        print("No es primo")
            
    return c   
```

```python
prim(3)
```
